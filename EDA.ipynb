### 3. ingr_B EDA

```python
# 히스토그램 그리기
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
sns.histplot(data['ingr_B'], bins=30, kde=True)
plt.title('Histogram')

# 밀도 추정 그래프 그리기
plt.subplot(1, 2, 2)
sns.kdeplot(data['ingr_B'])
plt.title('Kernel Density Estimation (KDE)')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

값의 편차가 너무 심하기 때문에 outlier를 제거하고 다시 확인

```python
# Z-스코어 계산
z_scores = (data['ingr_B'] - data['ingr_B'].mean()) / data['ingr_B'].std()

# Z-스코어 기준으로 이상치 제거 (예: Z-스코어가 3보다 큰 경우를 이상치로 판단)
filtered_data = data[z_scores < 3]  # 이상치를 제외한 데이터

# 이상치를 제외한 데이터로 히스토그램 그리기
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
sns.histplot(filtered_data['ingr_B'], bins=30)
plt.title('Histogram (Outliers Removed)')

# 밀도 추정 그래프 그리기
plt.subplot(1, 2, 2)
sns.kdeplot(filtered_data['ingr_B'])
plt.title('Kernel Density Estimation (KDE)')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

```python
print(f'Number of Outliers: {len(data[z_scores>3])}')
```

```python
# 이상치를 제외한 데이터로 히스토그램 그리기
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
sns.histplot(normal_data['ingr_B'], bins=30)
plt.title('Normal Histogram')

# 밀도 추정 그래프 그리기
plt.subplot(1, 2, 2)
sns.kdeplot(normal_data['ingr_B'])
plt.title('Kernel Density Estimation (KDE)')

# 그래프 보여주기
plt.tight_layout()
plt.show()
```

outlier 데이터가 비정상 데이터가 아닌 정상 데이터에서 나온거였다.
이렇다면 ingr_B는 항상 outlier를 정리하고 진행해도 괜찮을듯
